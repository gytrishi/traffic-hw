```python
import pandas as pd
```

```python
k=pd.read_excel('/content/traffic-hw/e1Detector_e2_0_0.xlsx',usecols="b:l",skiprows=1)
df=pd.DataFrame(k)
df
```

```python
import matplotlib.pyplot as plt
y= df['/interval/@flow']
x=df['/interval/@speed']
curve1= plt.plot(x,y)
```

```python
#curve fit
import numpy as np
import scipy.optimize
def parabola(x, a, b, c):
  return a*x**2 + b*x + c
fitting=scipy.optimize.curve_fit(parabola,df.density,df['/interval/@flow'],p0=[1,2,4])
ans,cov= fitting
a,b,c= ans
plt.errorbar(df.density,df['/interval/@flow'],fmt='r.',label="data")
plt.ylabel("flow")
plt.xlabel("density")
t= np.linspace(0,500)
plt.plot(t,parabola(t,a,b,c),label="fit line")
plt.legend()
```

```python
#curve fit
def triangle(x, a, b,c):
  return a * np.exp(-b * x) + c
fitting=scipy.optimize.curve_fit(triangle,df.density,df['/interval/@flow'],p0=[1,2,4])
ans,cov= fitting
a,b,c= ans
plt.errorbar(df.density,df['/interval/@flow'],fmt='r.',label="data")
plt.ylabel("flow")
plt.xlabel("density")
t= np.linspace(0,500)
plt.plot(t,triangle(t,a,b,c),label="fit line")
plt.legend()
```
